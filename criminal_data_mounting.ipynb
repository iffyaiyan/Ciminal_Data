```
# Trying to plot age category of the criminals

labels = df3['age_cat'].value_counts().index
values = df3['age_cat'].value_counts().values

colors = df3['age_cat']

fig = go.Figure(data=[go.Pie(labels = labels, values = values, textinfo="label+percent", marker=dict(colors=colors))])

fig.show()
```

from the above plot we can see that more than 50% infact 57% of the crimes are being commited by people in the age group of 25-45

```
# Trying to plot sex of the criminals

labels = df3['sex'].value_counts().index
values = df3['sex'].value_counts().values

colors = df3['sex']

fig = go.Figure(data=[go.Pie(labels = labels, values = values, textinfo="label+percent", marker=dict(colors=colors))])

fig.show()
```

```
# Trying to plot age category of the criminals

labels = df3['race'].value_counts().index
values = df3['race'].value_counts().values

colors = df3['race']

fig = go.Figure(data=[go.Pie(labels = labels, values = values, textinfo="label+percent", marker=dict(colors=colors))])

fig.show()
```

As per the data most of the criminals are the African-American

```
# Plotting the relation between the race and the sex of the criminals 

dxp.count(val='race', data=df3, normalize=True, split="sex", figsize=(4,3))
```

```
# Plotting the relation between the race and the age of the criminals 

dxp.count(val='race', data=df3, normalize=True, split="age_cat", figsize=(4,3))
```

From the above datasets we can infer out few of the most interesting things
1. African American have more proportional of criminal in the age group of (age less than 25) than the other race
2. Caucasian have more proportion of criminal in the age group of (age greater than 45) than any other race
3. Generally Native Americans are less indulged in crime as per the dataset

```
# Calculating the number of days between the day when the criminal was put in the jail and the days he was released


from datetime import datetime
date_format = "%Y%m/%d %H:%M:%S"
a = datetime.strptime('c_jail_in', date_format)
b = datetime.strptime('c_jail_out', date_format)
delta = b - a
print(delta.days) # that's it
```
